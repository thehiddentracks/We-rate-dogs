## Report: act_report

Within the "WeRateDogs" project I gathered various datasets based on the Twitter user "dog_rates" and his tweets about dog ratings. 

With the given data I was able to get an overview of the most often posted dogs, the favorite dogs amongst the account followers, and the overall ratings given. Let's start at the beginning.
All the tweets combined show an immensive amount of dog breeds rated by "dog_rates", but there seems to be favorites amongst the user. A well-owned first place goes to the Golden Retriever showing op the most in all the tweets created (see Figure 1). Also popular on the second and third place we find the Pembroke (a Corgi breed) and the Labrador Retriever.

###### Figure 1

But does this mean that the Golden Retriever is also the most popular breed amongst the account followers? Let's have a look at the favorite count per breed. In total sum, the Golden Retriever got the most "likes" from followers (see Table 1). But when we now look at the mean of the likes per breed (see Figure 2), his popularity gets shattered as breeds like the Irish Water Spaniel, the Saluki, or the Standard Poodle get way more likes by the audience than the Golden Retriever. The Golden Retriever isn't even in the Top-10 anymore and seems to look that popular in the first place just because he appears the most in the tweets.

The last finding involves the overall way of rating dogs by the account. The maximum of the scale of ratings (the "denominator") is set to 10. As we are talking about dogs and they are supposed to be sweet and fabulous all the time, the account owner gives a mean rating of 10.95 per dog and therefore exceeding the set denominator of 10 regularly. So while dogs get an fabulous rating on average, there are still outliers present with a top rating of 26 and 27 ("extraordinary dog") or even low rating of 2 ("bad dog"). Overall we can say that the account owner is a good person not trying to upset the dog owners except something is really off, but that seems to be rarely the case.
